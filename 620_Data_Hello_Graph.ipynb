```python
import networkx as nx
import matplotlib.pyplot as plt 
```

```python
G = nx.Graph()

G.add_edges_from([("Tyson", "Alex"),
                  ("Tyson", "Vanessa"),
                  ("Tyson", "Sergey"),
                  ("Tyson", "Fernando"),
                  ("Sergey", "Vanessa"),
                  ("Sergey", "Edmond"),
                  ("Sergey", "Gabriel"),
                  ("Alex", "Vanessa"),
                  ("Alex", "Fernando"),
                  ("Vanessa", "Edmond"),
                  ("Vanessa", "Fernando"),
                  ("Vanessa", "Gabriel"),
                  ("Edmond", "Gabriel"),
                  ("Fernando", "Gabriel"),
                  ("Fernando", "Patricia"),
                  ("Gabriel", "Patricia"),
                  ("Patricia", "Tony"),
                  ("Tony", "Troy")])

nx.draw(G, with_labels = True, font_weight="bold")
plt.show()
```
